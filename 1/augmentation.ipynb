```python
import numpy as np
import random
import itertools
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def load_audio_sec(file_path, sr=16000):
    data, _ = librosa.core.load(file_path, sr)
    if len(data) > sr:
        data = data[:sr]
    else:
        data = np.pad(data, pad_width=(0, max(0, sr - len(data))), mode="constant")
    return data

def plot_time_series(data):
    fig = plt.figure(figsize=(14, 8))
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
```

```python
data = load_audio_sec("p225_001.wav")
plot_time_series(data)
```

```python
#Hear it ! 
ipd.Audio(data, rate=16000)
```

```python
# Adding white noise 
wn = np.random.randn(len(data))
data_wn = data + 0.005*wn
plot_time_series(data_wn)
# Limited amplitude of the noise so we can still hear the word even with the noise, 
ipd.Audio(data_wn, rate=16000)
```

```python
# Shifting the sound
data_roll = np.roll(data, 8000)
plot_time_series(data_roll)
ipd.Audio(data_roll, rate=16000)
```

```python
data_stretch = librosa.effects.time_stretch(data, 0.8)
print("This makes the sound deeper but we can still hear 'off' ")
plot_time_series(data_stretch)
ipd.Audio(data_stretch, rate=16000)
```

```python
data_stretch =librosa.effects.time_stretch(data, 1.2)
print("Higher frequencies  ")
plot_time_series(data_stretch)
ipd.Audio(data_stretch, rate=16000)
```
